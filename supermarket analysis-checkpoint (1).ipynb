### Branch with the highest Gross income

```python
BG = df.groupby("Branch").sum()
BG.sort_values("Gross_income", ascending=False , 
inplace = True)
BG.head()
```

```python
plt.figure(dpi=80)
BG["Gross_income"].plot.pie(colors=["lightBlue","lightgreen","pink"],
shadow=True,explode=(0.,0.1,0.1),autopct=("%1.2f%%"),
title= "Branch with the highest Gross income \n 2019", fontsize = 8)
plt.legend()
```

### Which branch has the hishest and lowest ratings

```python
BR=df.groupby("Branch").sum()
BR.sort_values("Rating", ascending =False, 
inplace = True)
BR.head()
```

```python
BR["Rating"].plot.area(title = "Which branch has the hishest and lowest ratings \n 2019", color="lightgreen")
```

### Invoice_ID that has the heighest unit price in health and beauty

### Distribution of the gender (Male and female)

```python
plt.figure(figsize = (7,5))
df["Gender"].value_counts().plot(kind="barh")
plt.title("Gender distribution", fontsize = 15)
plt.show()
```

### What gender has the heighest total profit value

```python
GT=df.groupby("Gender").sum()
GT.sort_values("Total", ascending = False, inplace = True)
GT.head()
```

```python
GT["Total"].plot.bar(title ="Gender with the heighest Total profit in 2019", color= "orange")
```

### What payment method is most frequenly used in the year

### what product has the highest quantity

```python
PQ = df[["Product_line", "Quantity"]].groupby(["Product_line"],
as_index = False).sum().sort_values(by = "Quantity", ascending = False)
PQ
```

```python
plt.subplots(figsize =(17, 6))
plt.title("Product by Quantity", fontsize =15)
sns.barplot(x = "Product_line", y = "Quantity", data = PQ)
```
